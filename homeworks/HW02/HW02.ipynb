```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#2.3.1
data = pd.read_csv('S02-hw-dataset.csv', sep=',')
print(data.head())
print(data.info())
print(data.describe())
```

```python
#2.3.2
#1
for i in data.columns:
    print(data[i].isna().mean())
data['age'] = data['age'].fillna(int(data['age'].mean()))
#2 
print(data.duplicated())
data = data.drop_duplicates()
#3
print(data[data['user_id']  <= 0])
print(data[(data['age'] > 100) | (data['age'] < 0)])
print(data[(data['purchases'] < 0) | ((data['purchases'] == 0) & (data['revenue'] > 0)) | ((data['revenue'] == 0) & (data['purchases'] > 0))])
data = data[(data['user_id']  > 0) & ((data['age'] <=100) & (data['age'] >= 0)) & ((data['purchases'] >= 0)) &\
            ~((data['purchases'] == 0) & (data['revenue'] > 0)) & ~((data['revenue'] == 0) & (data['purchases'] > 0))]
#4
# 1 были обнаружены пропуски в столбце age
# 2 41 строка - дубликат
# 3 в строке 5 и возраст больше 100, и кол-во покупок - отрицательное число
# 4 есть также строки, в которых при кол-ве покупок > 0, доход равен 0
```

```python
#2.3.3
#1 
print(data['country'].value_counts())
#2
print(data.groupby('country').agg({'revenue':'median'}))
#3
age_bins = [0, 18, 25, 35, 45, 55, 65, 100]
age_labels = ['0-18', '19-25', '26-35', '36-45', '46-55', '56-65', '66+'] 
data['age_group'] = pd.cut(data['age'], bins=age_bins, labels=age_labels, include_lowest=True)
print(data.groupby('age_group', observed=False).agg({'revenue' : 'mean'}))
age_group_counts = data.groupby('age_group', observed=False)['age'].count()
print(age_group_counts)
# 4

# 1 больше всего пользователей из России, меньше всего из Китая.
# 2 больше всего доход от пользователей из Германии
# 3 самый большой доход получен от пользователей в возрастной группе от 0 до 18 лет. нет пользоваталей,
# совершивших покупки, в группе от 66 лет
# 4 больше всего пользователейй в группе от 26 до 35 лет
```

```python
# 2.4
# 1
fig3, ax3 = plt.subplots(figsize=(10, 6))
age_group_counts.plot(kind='hist', ax=ax3)
ax3.set_title("Распределение количества пользователей по возрастным группам")
ax3.set_xlabel("Возрастные группы")
ax3.set_ylabel("Количество пользователей")
ax3.grid(True, alpha=0.3)
plt.savefig(f'{'./figures'}/histogram.png', dpi=300, bbox_inches='tight')
plt.show()

# 2
fig2, ax2 = plt.subplots(figsize=(10, 6))
data.boxplot(column='revenue', by='age_group', ax=ax2)
plt.suptitle('')
ax2.set_title("Распределение потраченной суммы по возрастным группам")
ax2.set_ylabel("Потраченная сумма")
ax2.set_xlabel("Возрастные группы")
plt.savefig(f'{'./figures'}/boxplot.png', dpi=300, bbox_inches='tight')
plt.show()

# 3
fig1, ax1 = plt.subplots(figsize=(10, 6))
data.plot.scatter(x='purchases', y='revenue', ax=ax1)
ax1.set_xlabel("Количество покупок")
ax1.set_ylabel("Потраченная сумма")
ax1.set_title("Зависимость между количеством покупок и потраченной суммой")
plt.savefig(f'{'./figures'}/scatter_plot.png', dpi=300, bbox_inches='tight')
plt.show()
```
